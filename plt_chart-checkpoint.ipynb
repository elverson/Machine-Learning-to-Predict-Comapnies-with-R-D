```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
path = r'C:\\Users\\lixue\\OneDrive\\Desktop\\smu\\MQF\\Quantitative Research\\project_part2\\'
df_plt = pd.read_excel(path+"all_params_plt.xlsx")

df_plt['logistic precision_1']=df_plt['logistic precision_1'].astype(float)
df_plt['logistic_precision_0']=df_plt['logistic_precision_0'].astype(float)
df_plt['xgboost_precision_1']=df_plt['xgboost_precision_1'].astype(float)
df_plt['xg_boost_precision_0']=df_plt['xg_boost_precision_0'].astype(float)
df_plt['LGBM_precision_1']=df_plt['LGBM_precision_1'].astype(float)
df_plt['LGBM_precision_0']=df_plt['LGBM_precision_0'].astype(float)

df_plt['logistic recall_1']=df_plt['logistic recall_1'].astype(float)
df_plt['logistic_recall_0']=df_plt['logistic_recall_0'].astype(float)
df_plt['xgboost_recall_1']=df_plt['xgboost_recall_1'].astype(float)
df_plt['xg_boost_recall_0']=df_plt['xg_boost_recall_0'].astype(float)
df_plt['LGBM_recall_1']=df_plt['LGBM_recall_1'].astype(float)
df_plt['LGBM_recall_0']=df_plt['LGBM_recall_0'].astype(float)

df_plt['logistic_F1_score_1']=df_plt['logistic_F1_score_1'].astype(float)
df_plt['logistic_F1_score_0']=df_plt['logistic_F1_score_0'].astype(float)
df_plt['xgboost_F1_score_1']=df_plt['xgboost_F1_score_1'].astype(float)
df_plt['xg_boost_F1_score_0']=df_plt['xg_boost_F1_score_0'].astype(float)
df_plt['LGBM_F1_score_1']=df_plt['LGBM_F1_score_1'].astype(float)
df_plt['LGBM_F1_score_0']=df_plt['LGBM_F1_score_0'].astype(float)

df_plt['logistic_AUC']=df_plt['logistic_AUC'].astype(float)
df_plt['LGBM_AUC']=df_plt['LGBM_AUC'].astype(float)
df_plt['xgboost_AUC']=df_plt['xgboost_AUC'].astype(float)



```

```python
df_precision = pd.DataFrame(df_plt,columns=["logistic precision_1","logistic_precision_0","xgboost_precision_1",
                                            "xg_boost_precision_0","LGBM_precision_1","LGBM_precision_0"])



df_recall = pd.DataFrame(df_plt,columns=["logistic recall_1","logistic_recall_0","xgboost_recall_1",
                                            "xg_boost_recall_0","LGBM_recall_1","LGBM_recall_0"])



df_f1 = pd.DataFrame(df_plt,columns=["logistic_F1_score_1","logistic_F1_score_0","xgboost_F1_score_1",
                                            "xg_boost_F1_score_0","LGBM_F1_score_1","LGBM_F1_score_0"])



df_auc = pd.DataFrame(df_plt,columns=["logistic_AUC","xgboost_AUC","LGBM_AUC"])




df_precision['year'] = df_plt['Year_2']
df_recall['year'] = df_plt['Year_2']
df_f1['year'] = df_plt['Year_2']
df_auc['year'] = df_plt['Year_2']


```

```python
df_precision.plot(x="year", y=["logistic precision_1","logistic_precision_0","xgboost_precision_1",
                                            "xg_boost_precision_0","LGBM_precision_1","LGBM_precision_0"], 
                  kind="line", figsize=(12, 8), ylabel='Precision', title='Precision Comparison')

plt.savefig(path+'df_precision.jpg', format='jpg', dpi=400)

df_recall.plot(x="year", y=["logistic recall_1","logistic_recall_0","xgboost_recall_1",
                                            "xg_boost_recall_0","LGBM_recall_1","LGBM_recall_0"], 
                  kind="line", figsize=(12, 8), ylabel='Recall', title='Recall Comparison')

plt.savefig(path+'df_recall.jpg', format='jpg', dpi=400)

df_f1.plot(x="year", y=["logistic_F1_score_1","logistic_F1_score_0","xgboost_F1_score_1",
                                            "xg_boost_F1_score_0","LGBM_F1_score_1","LGBM_F1_score_0"], 
                  kind="line", figsize=(12, 8), ylabel='F1 Score', title='F1 Score Comparison')


plt.savefig(path+'df_f1.jpg', format='jpg', dpi=400)

df_auc.plot(x="year", y=["logistic_AUC","xgboost_AUC","LGBM_AUC"], 
                  kind="line", figsize=(12, 8), ylabel='AUC', title='AUC Comparison')


plt.savefig(path+'df_auc.jpg', format='jpg', dpi=400)
```
